# Titanic DataSet EDA

### Import  libraries

```python
# Import  libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the Titanic dataset

```python

# Load the Titanic dataset
data = pd.read_csv('C:\\Users\\Administrator\\Downloads\\tested.csv')
```

### Display first few rows of the dataset

```python
# Display first few rows of the dataset
print("Dataset preview:")
data.head()
```

### Data Cleaning

```python
# Data Cleaning
# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())
```

### Fill missing Age with median

```python

# Fill missing Age with median
data['Age'].fillna(data['Age'].median())
```

### Fill missing Embarked with the most common value (mode)

```python

# Fill missing Embarked with the most common value (mode)
# If there are missing values, fill them with the most common (mode) value in 'Embarked'
if data['Embarked'].isnull().sum() > 0:
    data['Embarked'].fillna(data['Embarked'].mode().iloc[0], inplace=True)

# Drop Cabin column due to many missing values
data.drop(columns=['Cabin'], inplace=True)
```

### Exploratory Data Analysis

```python
# Exploratory Data Analysis
plt.figure(figsize=(10, 6))

# 1. Survival Rate Count
plt.subplot(2, 2, 1)
sns.countplot(x='Survived', data=data)
plt.title('Survival Rate')
```

### 2. Survival Rate by Gender

```python
# 2. Survival Rate by Gender
plt.subplot(2, 2, 2)
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival Rate by Gender')
```

### 3. Age Distribution

```python
# 3. Age Distribution
plt.subplot(2, 2, 3)
sns.histplot(data['Age'], kde=True)
plt.title('Age Distribution')

```

### 4. Survival Rate by Class (Pclass)

```python
# 4. Survival Rate by Class (Pclass)
plt.subplot(2, 2, 4)
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title('Survival Rate by Class')

plt.tight_layout()
plt.show()
```

### Additional Analysis - Family Relationships

```python
# Additional Analysis - Family Relationships
plt.figure(figsize=(10, 6))

# 5. Survival Rate by Number of Siblings/Spouses (SibSp)
plt.subplot(1, 2, 1)
sns.countplot(x='SibSp', hue='Survived', data=data)
plt.title('Survival Rate by Siblings/Spouses')
```

### 6. Survival Rate by Number of Parents/Children (Parch)

```python
# 6. Survival Rate by Number of Parents/Children (Parch)
plt.subplot(1, 2, 2)
sns.countplot(x='Parch', hue='Survived', data=data)
plt.title('Survival Rate by Parents/Children')

plt.tight_layout()
plt.show()
```

### Fare Distribution and Embarked Analysis

```python

# Fare Distribution and Embarked Analysis
plt.figure(figsize=(10, 6))

# 7. Fare Distribution
plt.subplot(1, 2, 1)
sns.histplot(data['Fare'], kde=True)
plt.title('Fare Distribution')
```

### 8. Survival Rate by Embarked Port

```python

# 8. Survival Rate by Embarked Port
plt.subplot(1, 2, 2)
sns.countplot(x='Embarked', hue='Survived', data=data)
plt.title('Survival Rate by Embarked Port')

plt.tight_layout()
plt.show()
```

# Complete Project EDA

```python
# Expolratory Data Analysis
# Import  libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
data = pd.read_csv('C:\\Users\\Administrator\\Downloads\\tested.csv')

# Display first few rows of the dataset
print("Dataset preview:")
data.head()

# Data Cleaning
# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())

# Fill missing Age with median
data['Age'].fillna(data['Age'].median())

# Fill missing Embarked with the most common value (mode)
# If there are missing values, fill them with the most common (mode) value in 'Embarked'
if data['Embarked'].isnull().sum() > 0:
    data['Embarked'].fillna(data['Embarked'].mode().iloc[0], inplace=True)

# Drop Cabin column due to many missing values
data.drop(columns=['Cabin'], inplace=True)

# Exploratory Data Analysis
plt.figure(figsize=(10, 6))

# 1. Survival Rate Count
plt.subplot(2, 2, 1)
sns.countplot(x='Survived', data=data)
plt.title('Survival Rate')

# 2. Survival Rate by Gender
plt.subplot(2, 2, 2)
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival Rate by Gender')

# 3. Age Distribution
plt.subplot(2, 2, 3)
sns.histplot(data['Age'], kde=True)
plt.title('Age Distribution')

# 4. Survival Rate by Class (Pclass)
plt.subplot(2, 2, 4)
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title('Survival Rate by Class')

plt.tight_layout()
plt.show()

# Additional Analysis - Family Relationships
plt.figure(figsize=(10, 6))

# 5. Survival Rate by Number of Siblings/Spouses (SibSp)
plt.subplot(1, 2, 1)
sns.countplot(x='SibSp', hue='Survived', data=data)
plt.title('Survival Rate by Siblings/Spouses')

# 6. Survival Rate by Number of Parents/Children (Parch)
plt.subplot(1, 2, 2)
sns.countplot(x='Parch', hue='Survived', data=data)
plt.title('Survival Rate by Parents/Children')

plt.tight_layout()
plt.show()

# Fare Distribution and Embarked Analysis
plt.figure(figsize=(10, 6))

# 7. Fare Distribution
plt.subplot(1, 2, 1)
sns.histplot(data['Fare'], kde=True)
plt.title('Fare Distribution')

# 8. Survival Rate by Embarked Port
plt.subplot(1, 2, 2)
sns.countplot(x='Embarked', hue='Survived', data=data)
plt.title('Survival Rate by Embarked Port')

plt.tight_layout()
plt.show()

# Summary
print("\nSummary of findings based on survival rates, age distribution, fare, class, and family relationships.")
```
